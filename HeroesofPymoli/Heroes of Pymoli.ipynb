## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count = purchase_data_df["Item Name"].value_counts()
len(item_count)

av_price = purchase_data_df["Price"].mean()

av_age = purchase_data_df["Age"].mean()

num_pur = purchase_data_df["Purchase ID"].value_counts()
len(num_pur)

total_rev = purchase_data_df["Price"].sum()

summary_df = pd.DataFrame({
    "Number of Items": [len(item_count)],
    "Average Price": (round(av_price,2)),
    "Average Age": (round(av_age)),
    "Number of Purchases": (len(num_pur)),
    "Total Revenue": (total_rev)
})
summary_df["Average Price"] = summary_df["Average Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
summary_df
```
